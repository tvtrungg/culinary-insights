```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('ingredients.csv')
```

```python
df.head()
```

```python
# Group by 'url', and aggregate 'ingredients' into a list
# Remove comments to see the result
grouped_df = df.groupby(['url', 'country', 'title', 'details', 'steps', 'rating', 'nutrition', 'dish_type'])['ingredients'].apply(list).reset_index()
```

```python
grouped_df.head()
```

```python
# Save to csv and sort by 'country'
grouped_df.sort_values(by='country', inplace=True)
grouped_df.to_csv('cleaned_recipes.csv', index=False)
```

```python
print(grouped_df.isnull().sum()) # Check for missing values
```

```python
# Split the 'details' column
grouped_df['details'] = grouped_df['details'].apply(ast.literal_eval)  # Convert string to dictionary
grouped_df['Prep Time'] = grouped_df['details'].apply(lambda x: x.get('Prep Time:', 'N/A')) # Return N/A if key is not found
grouped_df['Cook Time'] = grouped_df['details'].apply(lambda x: x.get('Cook Time:', 'N/A'))
grouped_df['Total Time'] = grouped_df['details'].apply(lambda x: x.get('Total Time:', 'N/A'))
grouped_df['Servings'] = grouped_df['details'].apply(lambda x: x.get('Servings:', 'N/A'))

# Split the 'nutrition' column
grouped_df['nutrition'] = grouped_df['nutrition'].apply(ast.literal_eval)
grouped_df['Calories'] = grouped_df['nutrition'].apply(lambda x: x.get('Calories', 'N/A'))
grouped_df['Fat'] = grouped_df['nutrition'].apply(lambda x: x.get('Fat', 'N/A'))
grouped_df['Carbs'] = grouped_df['nutrition'].apply(lambda x: x.get('Carbs', 'N/A'))
grouped_df['Protein'] = grouped_df['nutrition'].apply(lambda x: x.get('Protein', 'N/A'))

# Drop the original columns
grouped_df = grouped_df.drop(columns=['details', 'nutrition'])
```

```python
# Chuyển cột dinh dưỡng thành số (loại bỏ 'g' trong Fat, Carbs, Protein)
grouped_df['Fat'] = grouped_df['Fat'].str.replace('g', '').astype(float, errors='ignore')
grouped_df['Carbs'] = grouped_df['Carbs'].str.replace('g', '').astype(float, errors='ignore')
grouped_df['Protein'] = grouped_df['Protein'].str.replace('g', '').astype(float, errors='ignore')
grouped_df['Calories'] = pd.to_numeric(grouped_df['Calories'], errors='coerce')
```

```python
grouped_df.to_csv('cleaned_recipes.csv', index=False)
```
