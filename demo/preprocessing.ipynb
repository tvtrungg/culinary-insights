```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

## Load data

```python
data = pd.read_csv("../data/cleaned_recipes.csv",index_col=0)
data.head()
```

```python
# Params declaration
col = data.columns 
for c in col : 
    if sum(data[c].isnull()) > len(data) * 0.7 : 
        data = data.drop(c,axis=1)
```

```python
data.drop([
    "url",
    "comments"
], axis=1, inplace=True)
```

```python
'g' in data['Fat'][0]
```

```python
col_edit = ["Fat", "Carbs", "Protein"]
for c in range(len(col_edit)) : 
    tmp = []
    for i in range(len(data)) : 
        if data[col_edit[c]][i] is not np.nan : 
            if 'g' in data[col_edit[c]][i] : 
                tmp.append(float(data[col_edit[c]][i].replace('g','')))
        else : 
            tmp.append(0)
    data[col_edit[c]] = tmp
```

```python
data
```

```python
t = []
for i in range(len(data)):
    if 'to' in data['servings'][i] : 
        lst = data['servings'][i].split('to')
        lst = [int(x) for x in lst]
        tmp = np.sum(lst) / len(lst)
        t.append(tmp)
    else :
        t.append(float(data['servings'][i]))

data['servings'] = t
```

```python
data.to_csv("../data/cleaned_recipes_v2.csv")
```
